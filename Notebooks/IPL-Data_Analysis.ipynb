```python
date=matches.match_date
key=[]
visited=[]
for i in date:
    if i in visited:
      number= visited.count(i)
      temp=i.split('-')
      temp1=temp[0]+temp[1]+temp[2]+'0'+str(number+1)
      visited.append(temp1)
      key.append(temp1)
    else:
      temp=i.split('-')
      temp1=temp[0]+temp[1]+temp[2]+'01'
      key.append(temp1)
      visited.append(i)
```

```python
matches['match_key']=key
```
